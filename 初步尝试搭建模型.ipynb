```python
%config InlineBackend.figure_format='svg'
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
from sklearn import preprocessing as ppc

with open('zhengqi_train.txt') as file:
    txt=file.readlines()
    #读取列标签存到columns
    #.split()函数默认去除空字符（包括空格和空行）
    columns=txt[0].split() 
    print('列标签为：',columns)
    #将txt的剩余内容按行读入data
    training_data=[]
    for line in txt[1:]:
        training_data.append(line.split())
        
##将zhengqi_train,zhengqi_test内的数据转化为.csv文件
with open('zhengqi_train.csv','w') as csvfile:
    writer=csv.writer(csvfile)
    #先写入columns_name
    writer.writerow(columns)
    #写入多行用writerows
    writer.writerows(training_data)

with open('zhengqi_test.txt') as file:
    txt=file.readlines()
    test_data=[]
    for line in txt[1:]:
        test_data.append(line.split())
    
with open('zhengqi_test.csv','w') as csvfile:
    writer=csv.writer(csvfile)
    writer.writerow(columns)
    writer.writerows(test_data)
    
```

```python
training_data=pd.read_csv('zhengqi_train.csv') #读取csv文件
print(training_data.head())
test_data=pd.read_csv('zhengqi_test.csv') #读取csv文件
print(test_data.head())
#print(training_data==test_data)
```

```python
X_train=training_data.iloc[:,:-1] #获取V0~V37的特征值
y_train=training_data.iloc[:,-1] #获取target的值
X_train=np.array(X_train) #shape=(2888,38)
y_train=np.array(y_train) #shape=(2888,)

##抽取数据选择数据标准化方法

spl_num=20 #抽取样本数目

X1=ppc.scale(X_train) #z-score处理
X2=(X_train-np.min(X_train,axis=0))/(np.max(X_train,axis=0)-np.min(X_train,axis=0)) #min-max处理
X3=np.arctan(X_train)*2/np.pi #反正切处理

plt.figure()
plt.xticks(np.arange(spl_num))
plt.xlabel('Index of sample features')
plt.ylabel('Value')
mean=[]
for j in range(spl_num):
    mean.append(np.mean(X_train,axis=0)[0])

plt.plot(np.arange(spl_num),mean,label='Mean') #均值横线
plt.plot(np.arange(spl_num),X_train[:spl_num,0],label='Raw',marker='o')
plt.plot(np.arange(spl_num),X1[:spl_num,0],label='Normalization',marker='o')
plt.plot(np.arange(spl_num),X2[:spl_num,0],label='Min-max scaling',marker='o')
plt.plot(np.arange(spl_num),X3[:spl_num,0],label='Arctan',marker='o')

plt.legend()
plt.show()
```

决定采用反正切处理方式
